## Bar Plots

```python
# Focus on distributions using categorical data in reference to one of the numerical columns

# Aggregate categorical data based on a function (mean is the default)
# بشكل افتراضي سيتم تجميع هذه البيانات بناء على المتوسط الحسابي , لكن يمكن تغييره عن طريق خاصية
# ال estimator
# Estimate total bill amount based on sex معناه المقدر اللي بقدر البيانات بتاء على اشي معين
# With estimator you can define functions to use other than the mean like those
# provided by NumPy : median, std, var, cov or make your own functions


sns.barplot(x='sex',y='total_bill',data = tips, estimator = np.median  )

# estemator = np.mean / np.median / np.std / np.var / np.cov /
# التقدير رح يكون مبين على القيمة اللي مش سترنج
# وحسب وين بحطها على اي محور رح يشتغل المقدر ويحط القيم اللي بده يقارن ويقدر على اساسها

# total_bill = numaric values > we Estimate total bill amount based on sex
```

# count plot

```python
# look like bar plot , but the estemator will count the total number of occurrences (التكرارات)

# A count plot is like a bar plot, but the estimator is counting
# the number of occurances


sns.countplot( x='sex' , data=tips)
# هون بدنا نعرف عدد الذكور والاناث من الداتا اللي اسمها tips
# sex = f (or) m
```

# box plot

```python
# A box plot allows you to compare different variables
# يسمح لنا بمفارة المتغيرات المختلفة
# فخو سوف يظهر الارباع الربعية للبيانات(الاول / الثاني / الثالث / الرابع )

# رؤية كيف يتغير الرجال والنساء فيما يتعلق باجمالي الفواتيرعلى مدار عدة ايام مختلفة

sns.set_style('darkgrid') # only styling

sns.boxenplot(x ='day' , y ='total_bill' , data= tips , hue='sex')

# standerd deviation = الانحراف المعياري
# الخط اللي بنص البوكس هو الانحراف المعياري
# حيث كل بوكس بمثل انحراف معياري واحد فقط

plt.legend(loc=0)
```

# violin plot

```python
# Violin Plot is a combination of the 1) boxplot and , 2) KDE , 1,2 هو مزيج بين الاثنين

'''
## While a box plot corresponds to data points, the violin plot uses the KDE estimation
    of the data points
     Split allows you to compare how the categories compare to each other

     يعني بوزع الرسمة بالاعتماد على ال kde
'''
sns.violinplot(x='day',y='total_bill',data=tips, hue='sex',split=True)

# split =true / false  :  تسمح لنا بمقارنة كيفية مقارنة الفئات مع بعضها البعض

# بالنهاية احنا رح نختار الرسم البياني الانسب النا على حسب نوع الداتا اللي عنا , يعني الشغلة مش عشوائية زي هلا

```

## Heat Maps

```python
plt.figure(figsize=(8,6)) # figure size
sns.set_context('paper', font_scale=1.4)

'''
To create a heatmap with data you must have data set up as a matrix
 where variables are on the columns and rows.

لعمل خريطة حرارية لازم ارتبها قبل اي اشي
واخليها بصيفة الاريه : بمتغيرات كاني عم بعمل ارييه ( صف , عمود )

يعني الاعمدة يلي عندي بالداتا بدي اخليها تتفهرس مش مع ارقام الاندكس
بدي اياها تتفهرس مع الاعمدة نفسها مرة تانية

يعني نستخدم الاعمدة كانهم صفوف

وفي طريقتين لعمل هاد الاشي  :

'''
#######################################################################

# الطريقةالاولى :

# [ corr() ]  function > استدعاء معامل الارتباط , حيث يبين الي مدى تاثير

#  استدعاء معامل الارتباط , حيث يبين الي مدى تاثير المتغير على النتائج اللي عندي

crash_matrix = crash.corr()
crash_matrix # عملنا ماتركس عشان الهيت ماب , وهاي القيم اللي بهاي الماتركس رح تستخدمها عشان نعمل الهيت ماب

sns.heatmap( crash_matrix , annot= True , cmap = 'crest' ) # cmap='Blues' >> ممكن احطها ازا بدي لون معين وممكن ما احطها , ممكن اطول باليت معين من موقع الماتبلوت الجاهزين


# annot : تعليقات توضيحية  , ورح تكون هون عبارة عن ارقام في وسط كل جزء من الخريطة


```

```python
#####################################################################

# الطريقة التانية :
#    نفس الاشي تقريبا لازم اعمل ماتركس
#  pivot table :  لكن بهاي الطريقة رح استخدم جدول البيفوت
#  new_df = df.pivot_table (index = ' ' , columns = ' ' , values = ' ' )

# index -> rows


flights = sns.load_dataset('flights')

flights = flights.pivot_table(index = 'month' , columns ='year' , values='passengers')

#flights
sns.heatmap(flights , cmap = 'Blues' , linecolor = 'white' , linewidths= 1 )

# linecolor -> الخطوط البيضة بين كل خلية في الهيت ماب ,
# linewidths -> لازم استعملها مع (linecolor) عشان عرض الخطوط يكون واضح
```

## Pair grid

- pair plots :
> برسملي كل العلاقات بين ال                
 ( columns ) اللي عندي

- pair grid
> بسمحلي اتحكم بنوع التمثيل البياني للرسمات
>> يعني انا احكيله شو نوع التمثيل البياني اللي بدي يمثلي كل العلاقات تبعت الداتا بين بعض

```python

iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(  iris , hue = 'species' ,
                        x_vars=["sepal_length", "sepal_width"] ,
                          y_vars=["petal_length", "petal_width"]   ) # انه رح نستعمل ال pairgrid
iris_g.map(plt.scatter)

# انه تمثيل الرسمات رح يكون بشكل ال scatter
# pairGrids allows us to control the type of plot for data representing , here we choosed the scatter plot .

iris_g.map(plt.scatter)

# Add a legend last
iris_g.add_legend()

```
